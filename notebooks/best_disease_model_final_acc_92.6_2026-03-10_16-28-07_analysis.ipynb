```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_theme(style='whitegrid')
```

```python
df = pd.read_csv('../reports/best_disease_model_final_acc_92.6_2026-03-10_16-28-07_detailed.csv')
df.head()
```

```python
print('--- Accuracy Overview ---')
print(df['Correct'].value_counts(normalize=True))

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Actual_Disease', hue='Correct')
plt.title('Correct vs Incorrect Predictions per Class')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import glob

# Model tag from your test.py run
model_tag = 'best_disease_model_final_acc_92.6_2026-03-10_16-28-07'
figures_path = '../figures'

# Try to find the figure with this model tag
expected_pattern = os.path.join(figures_path, f'{model_tag}-figure.png')
matching_files = glob.glob(expected_pattern)

if matching_files:
    img_path = matching_files[0]
    img = mpimg.imread(img_path)
    plt.figure(figsize=(12, 10))
    plt.imshow(img)
    plt.axis('off')
    plt.title(f'Confusion Matrix: {os.path.basename(img_path)}', fontsize=14, pad=20)
    plt.tight_layout()
    plt.show()
    print(f'✓ Displaying: {os.path.basename(img_path)}')
else:
    print(f'✗ No figure found matching: {expected_pattern}')
    print('\nAvailable figures:')
    if os.path.exists(figures_path):
        for f in os.listdir(figures_path):
            if f.endswith('.png'):
                print(f'  - {f}')
    else:
        print(f'Figures folder not found at: {figures_path}')
```
